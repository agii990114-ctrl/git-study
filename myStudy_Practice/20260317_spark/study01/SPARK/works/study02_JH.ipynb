```python
from sqlalchemy import create_engine
engine_mariadb = create_engine('mysql+pymysql://root:1234@192.168.0.204:3306/edu')
```

```python
from sqlalchemy import inspect
inspector = inspect(engine_mariadb)
tables = inspector.get_table_names()
print(tables)
```

```python
import pandas as pd
result = pd.read_sql_query("select * from edu.melon", engine_mariadb)
print(result)
```

```python
result.columns = ["id","code","title","album","img","cnt","regDate","modDate"]
result.head()
```

```python
result.title
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Study2_JH").master("spark://master:7077").getOrCreate()
```

```python
sDf = spark.createDataFrame(result)
```

```python
sDf.show()
```

```python
sDf.coalesce(1).write.format("com.databricks.spark.csv").option("header","true").save(path="/opt/spark/data/study2_JH.csv",format="csv",mode="overwrite")
```
