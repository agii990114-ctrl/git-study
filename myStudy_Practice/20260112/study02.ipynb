### 연산자
```
- + : 덧셈

- * : 곱샘
- % : 나머지
- / : 나눗셈
- // : 몫

```


### 비교연산자

```
== : 같다
!= : 다르다
< : 작다
> : 크다
<= : 작거나 같다
>= : 크거나 같다

**js에서의 === 는 없다**
```

### 논리연산자

- and print(1==1 and 1!=2) // true
- or print(1==1 or 1!=1) // true
- not print(not 1!=1) // true

````python
print(1==1 and 1!=2)
print(1==1 or 1!=1)
print(not 1!=1)
print(not 1!=1 or (1!=2 and 1!=1))
````

````python
#/ 8. 문자 F의 인덱스값을 찾으시오.
a = [
  ['A', 'B', 'C'],
  ['D', 'E', 'F'],
  ['G', 'H', 'I']
]
b = []
c = 0

for i in a:
    d = 0   
    for j in i:
        if j=='F':
            b=[c,d]
        d = d+1
    c = c +1
        
print(b)
````

````python
# 6) 다음 배열의 각 행 별 합계를 구하시오.
arr1 = [
  [9,4,6,7],
  [1,3,7,5],
  [5,8,2,4],
]

sum = []

for i in arr1:
    sumRow = 0
    for j in i:
        sumRow += j
    sum += [sumRow]
    
print(sum)
````
